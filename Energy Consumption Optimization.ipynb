```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# --- STEP 1: Medical Images ka Simulated Data banana ---
# Maan lete hain hamare paas 200 patients ke X-ray images hain, aur har image 64x64 pixels ki hai
np.random.seed(42)
num_images = 200
pixel_features = 64 * 64  # 4096 features per image

# Random pixel values generate kar rahe hain (0 se 255 ke beech me)
X_images = np.random.randint(0, 256, size=(num_images, pixel_features))

# Target: 0 matlab 'Normal', 1 matlab 'Disease Detected'
y_labels = np.random.choice([0, 1], size=num_images)

# --- STEP 2: Data ko Train aur Test me split karna ---
X_train, X_test, y_train, y_test = train_test_split(X_images, y_labels, test_size=0.2, random_state=42)

# --- STEP 3: Image Classification ke liye Model Train karna ---
# Hum RandomForest use kar rahe hain jo pixels ko bohat jaldi seekh leta hai
model_img = RandomForestClassifier(n_estimators=50, random_state=42)
model_img.fit(X_train, y_train)

# --- STEP 4: Predictions aur Accuracy Check karna ---
predictions = model_img.predict(X_test)
accuracy = accuracy_score(y_test, predictions)

print("--- Project 4: Medical Image Classification Complete! ---")
print(f"Image Classification Accuracy: {accuracy * 100:.2f}%")
print("\nDetailed Report:\n", classification_report(y_test, predictions, target_names=['Normal', 'Disease']))

# --- STEP 5: Ek Sample Image Visualise karna ---
# Ek random array ko 64x64 ke X-ray block me dekhne ke liye
sample_image = X_images[0].reshape(64, 64)
plt.figure(figsize=(4, 4))
plt.imshow(sample_image, cmap='bone') # 'bone' colormap X-ray jaisa look deta hai
plt.title("Sample Simulated Medical Scan")
plt.axis('off')
plt.show()
```
